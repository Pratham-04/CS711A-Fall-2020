```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

reward = np.genfromtxt("fin_r", delimiter = "\n")
rew = reward[0:50000]
rew = np.reshape(rew, [50,1000])

step = np.linspace(0,1,50)
# scat = []
# for i in range(len(step)):
#     for j in range(1000):
#         scat.append(step[i])
# plt.scatter(scat, reward, s = 0.2)

means = np.mean(rew, axis = 1)
maxI = list(means).index(np.max(means))
print(step[maxI])
plt.plot(step, means)
plt.xlabel("Epsilon (\u03B5)")
plt.ylabel("Cumulative Score\n(1000 iterations)")
plt.axvline(x = step[maxI], c = 'r', lw = 0.5, ls = '-.')
plt.axhline(y = np.max(means), c ='g', lw = 0.5, ls = '-.')
plt.annotate('('+str(np.round(step[maxI],2))+','+str(np.round(np.max(means),2))+')', (step[maxI]+0.01,np.max(means)+1))
plt.ylim([190, 230])
# plt.legend()
plt.show()
```
